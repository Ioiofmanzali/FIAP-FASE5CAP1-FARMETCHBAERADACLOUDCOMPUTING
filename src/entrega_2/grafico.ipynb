### Gráfico Comparativo  

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Variáveis e dados
localidades = ["São Paulo", "Virgínia do Norte"]
custos_mensais = [18.978, 13.974]
custos_anuais = [227.736, 167.688]
```

```python
x = np.arange(len(localidades))  
largura = 0.3  

fig, ax = plt.subplots(figsize=(8, 5))
barras_mensais = ax.bar(x - largura/2, custos_mensais, largura, label='Custo Mensal (USD)', color='royalblue')
barras_anuais = ax.bar(x + largura/2, custos_anuais, largura, label='Custo Anual (USD)', color='orange')


ax.set_xlabel('Localidade')
ax.set_ylabel('Custo (USD)')
ax.set_title('Comparação de Custos AWS')
ax.set_xticks(x)
ax.set_xticklabels(localidades)
ax.legend()

# Adicionando os valores nas barras
for barra in barras_mensais + barras_anuais:
    height = barra.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(barra.get_x() + barra.get_width() / 2, height),
                xytext=(0, 2),  
                textcoords="offset points",
                ha='center', va='bottom')

# Mostrar gráfico
plt.show()
```
